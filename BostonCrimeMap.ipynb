#Group Data

```python
crime_df['Lat'].dropna(inplace=True)
crime_df['Long'].dropna(inplace=True)
```

```python
crime_df
```

##Drop NaNs

```python
nan_locs = crime_df[crime_df['Lat'].isna()].index
nan_districts = crime_df[crime_df['DISTRICT'].isna()].index
crime_df.drop(nan_locs, inplace=True)
crime_df.drop(nan_districts, inplace=True)
```

#Crime Map Function

```python
def map_crimes(df, crime_type, district='All'):
  
  boston_districts = {'Downtown': 'A1', 'Charleston':'A15', 
    'East Boston':'A7', 'Roxbury':'B2',
    'Mattapan':'B3', 'South Boston':'C6',
    'Dorchester':'C11', 'South End':'D4',
    'Brighton':'D14', 'West Roxbury':'E5',
    'Jamaica Plain':'E13', 'Hyde Park':'E18'}

  if district != 'All':
    grouping = crime_df.groupby(['OFFENSE_CODE_GROUP', 'DISTRICT'])
    group = grouping.get_group((crime_type, boston_districts[district]))
  else: 
    grouping = crime_df.groupby('OFFENSE_CODE_GROUP')
    group = grouping.get_group(crime_type)

  map = folium.Map([42.3601,-71.0589], zoom_start=12)

  for crime in group.index:
    folium.Marker([group.loc[crime]['Lat'], group.loc[crime]['Long']], popup=group.loc[crime]['OFFENSE_DESCRIPTION']+'\n'+group.loc[crime]['Location']).add_to(map)

  return map
```

#Map of Homicides in Boston

```python
map_crimes(crime_df, 'Homicide')
```

#Map of Larcenies in Jamaica Plain

```python
map_crimes(crime_df, 'Larceny', 'Hyde Park')
```
