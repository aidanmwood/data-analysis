#Imports

```python
import pandas as pd
```

#Import Data

```python
url = 'https://raw.githubusercontent.com/bobg207/Honors_Data_Analysis/master/seattle_real_estate.csv'

data_df = pd.read_csv(url)
```

```python
data_df
```

#Normalize Data

```python
normal_df = pd.DataFrame()

normal_df['beds norm'] = (data_df.bedrooms - data_df.bedrooms.mean())/data_df.bedrooms.std()
normal_df['baths norm'] = (data_df.bathrooms - data_df.bathrooms.mean())/data_df.bathrooms.std()
normal_df['floors norm'] = (data_df.floors - data_df.floors.mean())/data_df.floors.std()
normal_df['sqft norm'] = (data_df.sqft_living - data_df.sqft_living.mean())/data_df.sqft_living.std()
normal_df['price norm'] = (data_df.price - data_df.price.mean())/data_df.price.std()

normal_df.sample(20)
```

#Find Closest Matches

## Ask user for parameters

```python
params = []

params.append(input('Enter the desired zip code: '))
params.append(float(input('Enter number of bedrooms: ')))
params.append(float(input('Enter number of bathrooms: ')))
params.append(float(input('Enter number of floors: ')))
params.append(float(input('Enter square feet of living space: ')))
params.append(float(input('Enter desired price ')))
```

```python
def best_matches(user_params, data):

  # Normalize Data
  normal_df = pd.DataFrame()

  normal_df['beds_norm'] = (data.bedrooms - data.bedrooms.mean())/data.bedrooms.std()
  normal_df['baths_norm'] = (data.bathrooms - data.bathrooms.mean())/data.bathrooms.std()
  normal_df['floors_norm'] = (data.floors - data.floors.mean())/data.floors.std()
  normal_df['sqft_norm'] = (data.sqft_living - data.sqft_living.mean())/data.sqft_living.std()
  normal_df['price_norm'] = (data.price - data.price.mean())/data.price.std()

  # Normalize User Data
  normal_params = []

  # Zip is not normalized, but included to keep indices consistent
  normal_params.append(user_params[0])

  normal_params.append((user_params[1] - data.bedrooms.mean())/data.bedrooms.std())
  normal_params.append((user_params[2] - data.bathrooms.mean())/data.bathrooms.std())
  normal_params.append((user_params[3] - data.floors.mean())/data.floors.std())
  normal_params.append((user_params[4] - data.sqft_living.mean())/data.sqft_living.std())
  normal_params.append((user_params[5] - data.price.mean())/data.price.std())

  zip_listings = {}

  # Get all listings in user selected zip
  for i in data.index:
    if data.statezip[i] == user_params[0]:
      # Find distance from user selections
      dist = (normal_df.beds_norm[i] - normal_params[1])**2 + (
          normal_df.baths_norm[i] - normal_params[2])**2 + (
              normal_df.floors_norm[i] - normal_params[3])**2 + (
                  normal_df.sqft_norm[i] - normal_params[4])**2 + (
                      normal_df.price_norm[i] - normal_params[5])**2

      zip_listings[data.index[i]] = dist
  
  return zip_listings
  
zip_listings = best_matches(params, data_df)
```

##Converts Listings dict to df

```python
zip_listings_df = pd.DataFrame.from_dict(zip_listings, orient='index')
```

##Sorts Listings based on mathematical distance from user's request

```python
zip_listings_df.sort_values(0, inplace=True)
```

#Prints Top 5 Listings

```python
count = 0
for i in zip_listings_df.index:
  if count < 5:
    print(data_df.iloc[i])
    count += 1
```
